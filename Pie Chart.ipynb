```python
import matplotlib.pyplot as plt
```

```python
classes=['python','ML','DL','NLP','Computer vision']
class1=[45,55,25,20,65]
```

```python
plt.pie([1])
plt.show()
```

# Labels Function

```python
plt.pie(class1,labels=classes)
plt.show()
```

# explode Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]
plt.pie(class1,labels=classes,explode=explode)
plt.show()
```

# Colors Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]
colors=['r','g','b','y','c']
plt.pie(class1,labels=classes,explode=explode,colors=colors)
plt.show()
```

# autopct Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]
colors=['r','g','b','y','c']
plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%')
plt.show()
```

# Shadow=True

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]
colors=['r','g','b','y','c']
plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True)
plt.show()
```

# radius  Function

# counterclock Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]

colors=['r','g','b','y','c']

textpros={"fontsize":12}

plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True,radius=2.0,
       startangle=90,pctdistance=0.6,labeldistance=0.3,counterclock=True)
plt.show()
```

# wedgeprops Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]

colors=['r','g','b','y','c']

textpros={"fontsize":12}

wedgeprops={'linewidth':10,'width':2,'edgecolor':"k"}

plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True,radius=2.0,
       startangle=90,pctdistance=0.6,labeldistance=0.3,counterclock=True)

plt.show()
```

# center Function

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]

colors=['r','g','b','y','c']

textpros={"fontsize":12}

wedgeprops={'linewidth':10,'width':2,'edgecolor':"k"}

plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True,radius=2.0,
       startangle=90,pctdistance=0.6,labeldistance=0.3,counterclock=True,center=(2,3))

plt.show()
```

# frame Function
Default True

```python
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]

colors=['r','g','b','y','c']

textpros={"fontsize":12}

wedgeprops={'linewidth':10,'width':2,'edgecolor':"k"}

plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True,radius=2.0,
       startangle=90,pctdistance=0.6,labeldistance=0.3,counterclock=True,frame=True)

plt.show()
```

# roratelabels Function

```python
plt.figure(figsize=(16,9))
classes=['python','ML','DL','NLP','Computer vision']
explode=[0.01,0.02,0.02,0.02,0.1]

colors=['r','g','b','y','c']

textpros={"fontsize":12}

wedgeprops={'linewidth':150,'width':4,'edgecolor':'r'}

plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',shadow=True,radius=2.0,
       startangle=90,pctdistance=0.6,labeldistance=1.3,counterclock=True,rotatelabels=True)

plt.legend(loc=0)

plt.show()
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
classes=['python','ML','DL','NLP','CV']
class1=[30,10,20,25,10]
class2=[40,10,30,35,10]
class3=[50,60,70,85,75]
```

```python
plt.bar(classes,class1,color='r')
plt.show()
```

# Horizontal chart

```python
plt.barh(classes,class1)
plt.show()
```

# color Function

```python
plt.bar(classes,class1,color='g')
plt.show()
```

# width Function

```python
plt.bar(classes,class1,width=0.4)
plt.show()
```

# align Function

'center', 'edge'

```python
plt.bar(classes,class1,align='center')
plt.show()
```

```python
plt.bar(classes,class1,align='edge')
plt.show()
```

# Edgecolor

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='y')
plt.show()
plt.show()
```

# alpha Function

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='y',alpha=0.9)
plt.show()
plt.show()
```

# linestyle Function

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--')
plt.show()
plt.show()
```

# label Function

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--',label='Students Marks') 

plt.xlabel("Subject",fontsize=15)
plt.ylabel('marks',fontsize=15)
plt.grid(True)
plt.title('Marks distribution',fontsize=15)
plt.show()
```

# visiable Function

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--',label='Students Marks'
       ,visible=False) 

plt.xlabel("Subject",fontsize=15)
plt.ylabel('marks',fontsize=15)
plt.grid(True)
plt.title('Marks distribution',fontsize=15)
plt.show()
```

```python

plt.bar(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--',label='Students Marks') 

plt.xlabel("Subject",fontsize=15)
plt.ylabel('marks',fontsize=15)
plt.grid(True)
plt.title('Marks distribution',fontsize=15)
plt.show()
```

```python
plt.barh(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--',label='Students Marks') 
plt.xlabel("Subject",fontsize=15)
plt.ylabel('marks',fontsize=15)
plt.grid(True)
plt.title('Marks distribution',fontsize=15)
plt.show()
```

```python
plt.bar(classes,class1,align='center',linewidth=2,edgecolor='r',alpha=0.9,linestyle='--',label='Students Marks',color='r') 
plt.bar(classes,class2,align='center',linewidth=2,edgecolor='y',alpha=0.9,linestyle='--',label='Students Marks',color='y')
plt.bar(classes,class3,align='center',linewidth=2,edgecolor='k',alpha=0.9,linestyle='--',label='Students Marks',color='k')
plt.xlabel("Subject",fontsize=15)
plt.ylabel('marks',fontsize=15)
plt.grid(True)
plt.title('Marks distribution',fontsize=15)
plt.show()
```

```python
class_index=np.arange(len(classes))
width=5
plt.figure( figsize=(16,9))
plt.bar(class_index,class1,align='center',linewidth=2,edgecolor='r',alpha=0.2,linestyle='--',label='Students Marks',color='r') 
plt.bar(class_index+width,class2,align='center',linewidth=2,edgecolor='y',alpha=0.9,linestyle='--',label='Students Marks',color='r')
plt.bar(class_index+width+width,class3,align='center',linewidth=2,edgecolor='k',alpha=0.9,linestyle='--',label='Students Marks',color='g')
plt.xlabel("Subject",fontsize=15)

plt.ylabel('marks',fontsize=15)
plt.grid(True,color='r',linestyle='--',linewidth=1)
plt.title('Marks distribution',fontsize=15)
plt.show()  
```

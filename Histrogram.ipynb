```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
ml_student=np.random.randint(18,50,(100))
```

```python
pl_student=np.random.randint(15,40,(100))
```

```python
print(ml_student)
print("\n")
print(pl_student)
```

# Histrogram Graph 

```python
bins=[15,20,25,30,40,45,50]
plt.figure(figsize=(16,9))
plt.hist(ml_student,bins,label='Ml Students',color='r',)
plt.axis([0,55,0,45])
plt.xlabel("Ml_student",fontsize=12)
plt.ylabel("No of students",fontsize=12)
plt.title("Students age Data Set",fontsize=15)
plt.grid(True,c='r',linestyle='--')
plt.legend(loc=3)
plt.show()
```

# rwidth Function

```python
bins=[15,20,25,30,40,45,50]
plt.figure(figsize=(16,9))
plt.hist(ml_student,bins,label='Ml Students',color='b',rwidth=0.8)
plt.axis([0,55,0,45])
plt.xlabel("Ml_student",fontsize=12)
plt.ylabel("No of students",fontsize=12)
plt.title("Students age Data Set",fontsize=15)
plt.grid(True,c='r',linestyle='--')
plt.legend(loc=0)
plt.show()
```

# Histtype

```python
bins=[15,20,25,30,40,45,50]
plt.figure(figsize=(16,9))
plt.hist(ml_student,bins,label='Ml Students',color='b',rwidth=0.8,histtype='barstacked')
plt.axis([0,55,0,45])
plt.xlabel("Ml_student",fontsize=12)
plt.ylabel("No of students",fontsize=12)
plt.title("Students age Data Set",fontsize=15)
plt.grid(True,c='r',linestyle='--')
plt.legend(loc=0)
plt.show()
```
